```python
import matplotlib.pyplot as plt
```

```python
from Pills.MaskedSequence import MaskedSequence
```

```python
def visualize(views):
  height = len(views)
  width = len(views[0])
  fig, axes = plt.subplots(height, width, figsize=(width*10, height*10))

  for row, view in enumerate(views):
    for col, (img, title) in enumerate(view):
      axes[row][col].imshow(img)
      axes[row][col].set_title(title)
      axes[row][col].axis('off')

  plt.show()
```

```python
BATCH_SIZE = 2
INPUT_SHAPE = (256, 256, 3)
```

```python
masked_sequence = MaskedSequence(100_000_000, BATCH_SIZE, INPUT_SHAPE[:2])
```

```python
idx = 0
```

```python
X, Y = masked_sequence[idx]
idx += 1

view_lines = []
for img, masked  in zip(X, Y):
  view_lines.append([(img, "masked"), (masked, "img")])

visualize(view_lines)
```
